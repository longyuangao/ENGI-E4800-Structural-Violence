```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import scipy.stats as stats
import datetime
import geocoder
```

```python
superfund = pd.ExcelFile('..\..\Local\superfund.xlsx')
print(superfund.sheet_names)
df=pd.read_excel('..\..\Local\superfund.xlsx')
for items in superfund.sheet_names[1:]:
    dfnew=pd.read_excel("..\..\Local\superfund.xlsx",sheet_name=items)
    df=pd.concat([df,dfnew])
superfund_df = df
```

```python
superfund_df["city_state"]=superfund_df.apply(lambda x: x["site_city_name"]+", "+x["rstate_code"]+", USA",axis=1)
```

```python
city_names = pd.DataFrame(superfund_df["city_state"].unique(),columns=["city_state"])
```

```python
def city_to_county(df:pd.DataFrame):
    my_key = input("API KEY:")
    confirm = input("Request Size:"+str(df.shape[0])+". Confirm?")
    if confirm == "yes":
        df["result"]=df["city_state"].apply(lambda x: geocoder.google(x, key=my_key))
        return df
    else: 
        return None
```

```python
test_df = city_names.head(3)
result=city_to_county(city_names)
```
